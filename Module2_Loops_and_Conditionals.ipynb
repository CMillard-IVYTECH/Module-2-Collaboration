```python
# =========================
# 4.1 - Conditional Logic
# =========================

secret = 7
guess = 3

if guess < secret:
    print("too low")
elif guess > secret:
    print("too high")
else:
    print("just right")
```

```python
# =========================
# 4.2 - Small and Green
# =========================

small = True
green = False

if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif not small and green:
    print("watermelon")
else:
    print("pumpkin")

```

```python
# =========================
# 6.1 - For Loop List
# =========================

for i in [3, 2, 1, 0]:
    print(i)
```

```python
# =========================
# 6.2 - While Loop
# =========================

guess_me = 7
number = 1

while True:
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break

    number += 1
```

```python
# =========================
# 6.3 - For Loop Range
# =========================

guess_me = 5

for number in range(10):
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break
```
